# beatifulSoup ÎùºÏù¥Î∏åÎü¨Î¶¨
- reqÎ°ú Î∞õÏïÑÏò® ÌÖçÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞Î•º Ïª¥Ïù¥ Ïù¥Ìï¥ Ìï†Ïàò ÏûàÍ≤å htmlÎ°ú Î≥ÄÌôò ÏãúÏºúÏ£ºÎäî ÎùºÏù¥Î∏åÎü¨Î¶¨

```python
# ÏûÑÌè¥Ìä∏ Ìï†Ïãú Ï†ÄÎèôÏôÑÏÑ±Ïãú stoneÏÇ¨Ïö© ÏïàÌï®
# bs4Ìè¥ÎçîÎ°úÎ∂ÄÌÑ∞ ÏûÑÌè¥Ìä∏ÌïúÎã§.
from bs4 import BeautifulSoup as bs
```

```python
# bsÏùÑ Ïù¥Ïö©Ìï¥ÏÑú reqÏùò ÌÖçÏä§Ìä∏ Ï†ïÎ≥¥Î•º htmlÎ°ú Î≥ÄÌôò
# Î¨∏ÏûêÏó¥Ïù¥ ÏïÑÎãàÎã§ Îî∞Ïò¥Ìëú ÏóÜÎã§ --> Ï¶â Ïª¥Ïù¥ Ïù¥Ìï¥ÌïòÎäî htmÏΩîÎìú -->Ï¶â Ïª¥Ïóê Î™ÖÎ†πÌï¥ÏÑú ÌäπÏ†ï Îç∞Ïù¥ÌÑ∞Î•º ÏàòÏßëÌï†Ïàò ÏûàÍ≤åÎêúÎã§.
# Ïª¥ÏùÄ ÌÉúÍ∑∏Î°úÎßå Ï†úÎ™©Ïù¥ÎÇò Î¨∏Îã®Ïù∏ÏßÄÎ•º ÏïåÏàò ÏûàÎã§
# Í∑∏ ÏΩîÎìúÏùÑ Ïù¥Ïö©Ìï¥ Ïö∞Î¶¨Îäî Ïª¥Í≥º ÌÜµÏã†ÌïúÎã§

soup = bs(res.text,"lxml")
soup
```

```python
# Ïª¥ÏóêÍ≤å soup Îç∞Ïù¥ÌÑ∞ = Ï¶â htmlÏóêÏÑú ÎÇ¥Í∞Ä ÌïÑÏöîÌïú ÌÉúÍ∑∏Îì§Îßå Í∞ÄÏ†∏Ïò§ÎùºÍ≥† Î™ÖÎ†πÌï¥Î≥¥Ïûê
# Î™®Îì† a ÌÉúÍ∑∏Î•º Í∞ÄÏ†∏ÏôÄ Î≥¥Ïûê
# soupÎç∞Ïù¥ÌÑ∞ÏóêÏÑú ÌïÑÏöîÌïú Îç∞Ïù¥ÌÑ∞Î•º Ï∂îÏ∂ú -> soup.select()
# selectÌï®ÏàòÎäî Î¶¨ÌÑ¥ Îç∞Ïù¥ÌÑ∞Í∞Ä Î¶¨Ïä§Ìä∏Î°ú ÎÑòÏñ¥Ïò®Îã§ --> Í≥†Î°ú Ïù∏Îç±Ïä§,Î∞òÎ≥µÎ¨∏ ÌôúÏö©Ìï† Ïàò ÏûàÎã§
# select_oneÌï®ÏàòÎèÑ ÏûàÎäîÎç∞ --> ÌïúÍ∞úÎßå Î¶¨ÌÑ¥ --> Î¶¨Ïä§Ìä∏Ïùò Ï≤´Î≤àÏß∏Îßå Î¶¨ÌÑ¥ --> ÏÇ¨Ïö© Ïûò ÏïàÌï®

# Ïó¨Í∏∞ÏÑúÎäî ÌÅ¥ÎûòÏä§ ÏÑ†ÌÉùÏûêÏù¥ÎØÄÎ°ú .Ï∞çÎäîÎã§ news_titÏùÄ ÌÅ¥ÎûòÏä§ Ïù¥Î¶ÑÏù¥Îã§ 
# aÌÉúÍ∑∏ Ï§ë news_tit ÎùºÎäî ÌÅ¥ÎûòÏä§Ïùò aÌÉúÍ∑∏Îßå ÏÑ†ÌÉùÌïòÍ≤†Îã§Îäî ÏùòÎØ∏
title = soup.select("a.news_tit")
title
# aÌÉúÍ∑∏ Ï§ë ÏÑ†ÌÉùÏûêÎ•º ÏïåÏïÑÏïº ÌïúÎã§ --> Ïó¨Í∏∞ÏÑúÎäî ÏÑ†ÌÉùÏûêÎ°ú ÌÅ¥ÎûòÏä§Í∞Ä Ï°¥Ïû¨ÌïúÎã§. Í≥†Î°ú a.news_titÎ•º Ï†ÅÎäîÎã§.
# Í∞úÎ∞úÏûêÎèÑÍµ¨ ÏÇ¨Ïö© -> ÏõêÌïòÎäî Î∂ÄÎ∂ÑÏùÑ Ïö∞ ÌÅ¥Î¶≠ÌõÑ Í≤ÄÏÇ¨Î•º ÎàÑÎ•¥Î©¥ Ìï¥Îãπ ÌÉúÍ∑∏Î•º Î≥ºÏàò ÏûàÎã§
```

```python
# titleÏóêÏÑú Í∏ÄÏûêÏ†ïÎ≥¥Îßå Ï∂úÎ†• --> .textÎ•º ÏÇ¨Ïö©
# news_tit ÌÅ¥ÎûòÏä§Ïù∏ aÌÉúÍ∑∏ÏóêÏÑú 1Î≤à Ïù∏Îç±Ïä§ text Í∞ÄÏ†∏Ïò§Í∏∞ --> []Ïù∏Îç±Ïã±Ïù¥ Í∞ÄÎä•Ìïú Ïù¥Ïú†Îäî Î∞îÎ°ú selectÌï®ÏàòÎ•º ÏÇ¨Ïö©Ìïú titleÎ≥ÄÏàòÎäî Î¶¨Ïä§Ìä∏ ÌÉÄÏûÖÏù¥ÎØÄÎ°ú
# Ïûò Î≥¥Î©¥ [aÌÉúÍ∑∏,aÌÉúÍ∑∏,....] --> 1Î≤àÏß∏ Ïù∏Îç±Ïä§ aÌÉúÍ∑∏Î•º Í∞ÄÏ†∏Ïò§Îäî Í≤å ÏïÑÎãàÍ≥† Í±∞Í∏∞ÏÑú textÎßå Í∞ÄÏ†∏Ïò§Îäî Í≤ÉÏù¥Îã§.
# ÏúÑÏóêÏÑú res.textÏ≤òÎüº~ texÎßå Í∞ÄÏ†∏Ïò®Îã§ --> urlÏöîÏ≤≠Ìïú ÏùëÎãµÏù¥ ÎîïÏÖîÎÑàÎ¶¨ ÌÉÄÏûÖÏù∏Îç∞ Í±∞Í∏∞ÏÑú Î∞∏Î•òÎ•º Í∞ÄÏ†∏Ïò§Í∏∞ ÏúÑÌï¥ res[key] Í∞Ä ÏïÑÎãå  res.keyÎ°ú Ï†ëÍ∑ºÌïúÎã§
# res.keyÏùò keyÍ∞Ä Î∞îÎ°ú textÏù¥Îã§
# title[1]ÏùÄ Î¶¨Ïä§Ìä∏Ïùò Ïù∏Îç±Ïä§Ïù∏Îç∞ Í±∞Í∏∞ÏÑú textÎßå Í∞ÄÏ†∏Ïò®Îã§ ÏïÑÎ¨¥Ìäº .text Îäî Î¨∏ÏûêÏó¥Îßå Í∞ÄÏ†∏Ïò®Îã§Í≥† ÏÉùÍ∞ÅÌïòÏûê
title[1].text 
```

```python
# Î™®Îì† Í∏ÄÏûê Ï†ïÎ≥¥Îßå Ï∂úÎ†•
# Î∞òÎ≥µÎ¨∏ÏùÑ ÌÜµÌï¥ Î™®Îì† a ÌÉúÍ∑∏Ïùò Î¨∏ÏûêÏó¥Îßå Í∞ÄÏ†∏Ïò§Ïûê --> Îç∞Ïù¥ÌÑ∞Ïùò Í∞ØÏàòÍ∞Ä Ï†ïÌï¥ÏßÑ Í≤ΩÏö∞ forÎ¨∏ ÎØ∏Ï†ïÏù¥Î©¥ whileÎ¨∏
# iÏùò Î≤îÏúÑÎäî 0Î∂ÄÌÑ∞ ~ len(title)=Í∏∏Ïù¥ÏóêÏÑú -1
for i in range(len(title)) :
    print(title[i].text)
```

```python
# i = titleÏïàÏùò aÌÉúÍ∑∏
for i in title :
    print(i.text)
```
