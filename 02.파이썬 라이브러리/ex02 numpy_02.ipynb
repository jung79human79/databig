### 블리언 인덱싱
- 배열안에서 조건에 만족하는 참인 값들만 추출하는 인덱싱 방법

```python
# 블리언 데이터 만들기 --> 데이터를 참/거짓 형태로 바꿔주자
rs = (bmi >=23) & (bmi <25)
rs
```

```python
# 블리언 인덱싱 진행하기
bmi[rs]
```

```python
# 변수생성 없이 바로 진행
bmi[(bmi >=23) & (bmi <25)]
```

- 기존에 인덱싱 진행하던 방법 : 인덱스 번호를 이용했다
- 블리언 인덱싱 진행 방법 : 인덱스 번호가 들어가는 곳에 조건식을 걸어 참/거짓으로 이루어진 데이터를 넣어준다
    - 블리언 인덱싱을 진행하면 값이 참인 데이터만 가지고 온다

- 블리언 인덱싱 이해하기

```python
# size=() --> 랜덤정수 numpy배열생성
rd_int= np.random.randint(1,101, size =(10,10))
rd_int
```

```python
# 위의 데이터에서 2의 배수이자 5의 배수인 데이터만 가져오자
# 블리언 데이터 만들기 (참/거짓)
# 조건확인 --> 2의 배수이자 5의 배수인지 확인하자
(rd_int%2==0) & (rd_int%5==0)
```

```python
# 위에서 만든 블리리언 데이터를 활용해서 인덱싱을 진행하자 --> []안에 넣기
print("2의 배수이자 5의 배수 : \n" , rd_int[(rd_int%2==0) & (rd_int%5==0)]) # 변수생성 안했다면 바로 조건식을 넣어준다
```

```python
# 배열내 모든 요소 갯수 
rd_int[(rd_int >=50) & (rd_int <80)].size  
```

###  최종실습
- 영화 데이터 평점 분석

```python
# 데이터 불러오기
data = np.genfromtxt("./data/ratings.dat",delimiter="::").astype("int64")
data
```

```python
# user_id , item_id  rating(평점) , timestamp
```

```python
# 배열의 차원 크기,전체요소 갯수 
print('차원:',data.ndim)
print('크기:',data.shape)
print('갯수:',data.size)
print('타입:',data.dtype)
```

```python
# 전체 평점 평균 구하기
# 평균--> rating(몇번 열인지 잘 고민)
np.mean(data[:,2]) 
data[:,2].mean()
```

```python
# np.unique()--> 데이터의 고유값을 구하는 함수
user_id = np.unique(data[:,0])
print(user_id)
print(user_id.shape)
user_id
```

```python
# 사용자 아이디만 블리언 인덱싱하기
# 아이디가 1번인 사람만 추출해보자
# 사용자 아이디가 1인사람을 가져와보자
# 우선 사용자 id에 접근을 한 후 아이디가 1인 불리언 데이터 만들기
# data[:, 0] == 1
# 위에서 만든 불리언 데이터로 불리언 인덱싱 진행하기
data[data[ : , 0 ] == 1]
```

```python
data[data[ : , 0 ] == 1]
```

```python
# 위에서 구한 사용자 아이디가 1인 데이터의 평점 평균 구하기
# 접근을 할때도 전체 행에 접근하고 평점 열에 접근 할 것!
user_one_mean = data[data[ : , 0 ] == 1][:, 2].mean()
user_one_mean
```

```python
# 사용자별 평균 평점 구하기 1단계
# 아이디가 1인 사람부터 6040번까지 반복문을 돌려 아이디별 평점 평균을 구하자
# 구한 평균은 데이터를 묶어주는 리스트를 통해 하나의 자료구조로 관리하자

rating_list = [] # 아이디 별 평균을 저장하는 빈 리스트

for i in user_id : # user_id만큼 반복문을 돌려주세요
    rs = data[data[ : , 0 ] == i][:, 2].mean() # 숫자를 바꾸면서 아이디별 평점 평균 구하기
    rating_list.append(rs) # 구한 평점 평균을 빈 리스트에 쌓아주세요
rating_list
```

```python
# 사용자별 평균 평점 구하기 2단계
# 아이디와 평점 평균 합쳐주기
# 우선 평점 평균은 리스트 형태
# 유저 아이디는 Numpy 배열
# 원활하게 값이 합쳐지도록 자료 형태를 통일 시켜주자
user_mean_arr = np.array(rating_list)
user_mean_arr
```

```python
# 사용자별 평균 평점 구하기 3단계
# 유저 아이디와 유저별 평균 평점이 합쳐진 데이터를 보관하기 위한 빈 리스트
user_mean_list = []

for i in user_id : # user_id만큼 반복문 돌리기
    # rating_list의 인덱스 번호는 0번부터 시작
    # user_id는 1번부터 시작 
    # 데이터를 처음부터 가져오기 위해 rating_list 인덱스 번호에 -1을 해줌
    # 1,2,3,4(user_id)...에 -1 -> 0,1,2,3...
    user_mean_list.append([i, rating_list[i-1]])
    
user_mean_list
```

```python
# 사용자별 평균 평점 구하기 4단계 = result
# 불리언 인덱싱을 진행하기 위해서 넘파이 배열로 변환
# 불리언 인덱싱은 리스트 / 튜플을 지원하지 않음
result = np.array(user_mean_list)
result
# 값이 이상하게 나오는 이유는 소수점 자리수가 비트를 초과함
# 지수형태로 나오는 것
# 형태는 이상하지만 원래 값 정보는 유지하고 있음
```

```python
# 평균 평점이 4점 이상인 사용자를 추려내기(정답 : 1544명)
# 평균 평점이 4점 이상인 사람만 추려내는 불리언 인덱싱
boolean_4_up = result[result[:, 1] >=4]
boolean_4_up
```

```python
# 사람의 수를 알아보기 위해 size를 찍어봄
boolean_4_up.size
# size는 전체 요소 갯수이기 때문에 1544 * 2(행의 갯수 * 열의 갯수)
```

```python
# 정확한 집계를 위해 하나의 열을 뽑아서 요소 갯수를 확인
boolean_4_up[:, 0].size
```

```python
# 다른 방법으로는 배열의 길이를 체크하는 방법이 있음
len(boolean_4_up)
```
