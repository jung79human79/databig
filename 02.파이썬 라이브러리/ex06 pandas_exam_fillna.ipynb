```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
```

```python
test1 = pd.DataFrame({'id':[1,2,3,4,5],'min':[60,80,70,90,85]},index=['1과목','2과목','3과목','4과목','5과목'])
test2 = pd.DataFrame({'id':[1,2,3,4,5],'final':[70,83,65,95,80]},index=['1과목','2과목','3과목','4과목','5과목'])
test_fill=pd.concat([test1, test2])
test_fill
```

```python
# fillna()함수 --> fill()함수는 없다
# final컬럼에 nan값을 85.0으로 채운다 
# 동일 컬럼명이므로 덮어쓰기
test_fill['final'] = test_fill['final'].fillna(85.0)
test_fill
```

```python
# 시리즈는  프레임의 컬럼이 된다
# 동일 이름의 컬럼을 파생변수로 해서 거기에 값을 넣어주면 결국 컬럼의 데이터를 수정 할수 있다
# 다른 이름의 컬럼을 파생변수로 하면 새로운 컬럼이 추가 되는 것이다 마치 insert 혹은 append
# del을 써서 컬럼를 삭제 할수 있다
test_fill['new'] = pd.Series([1,2,3,4,5],index=['1과목','2과목','3과목','4과목','5과목'])
test_fill
# 프레임에서 인덱스명을 따로 지정 안하면 기본값인 0부터~ 시작이므로 시리즈 생성시에도 인덱스명을 안적어도 된다
# 시리즈 생성시 인덱스 기본값이 0부터~ 시작이므로 동일하기 때문에~
# 그러나 프레임에서 인덱스명을 지정하면 시리즈 생성시에도 즉 새로운 컬럼을 만들때도 역시 프레임에서 새로 지은 인덱스명
# 같이 적어줘야 한다
```

```python
# min컬럼에 nan값을 45.0으로 채운다 
test_fill['min'] = test_fill['min'].fillna(45.0)
test_fill
```
