```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
data = pd.read_csv('./data/Traffic_Accident_2017.csv', encoding = 'euc-kr')
data
```

```python
data['요일'].unique()
```

```python
# 1. 조건 사고유형 대분류가 차대사람
aa = data[data['사고유형_대분류'] == '차대사람']
aa
```

```python
#  2.조건 발생시간대가 야간 인 경우

bb = aa[aa['주야'] == '야간']
bb
```

```python
bb.groupby('요일').agg(사고건수=('사상자수','count'))
```

```python
cc=bb.groupby('요일').agg(사고건수=('사상자수','count'))
dd=cc.reindex(index=['월','화','수','목','금','토','일'])
dd
```

```python
x = dd.index
y = dd.values
```

```python
plt.figure(figsize=(15,8))
plt.plot(x,y)
plt.xlabel('요일')
plt.ylabel('사고건수',rotation = 0)
plt.title('요일별 야간 차대사람 교통사고 건수')
plt.show()
```
