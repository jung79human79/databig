### barplot

```python
# 데이터 준비하여 막대 그래프 그려보기
np.random.seed(6)
bar_x = np.arange(1,4)
bar_y = np.random.randint(50,100,size=3)
# plt.bar(x축,y축)
plt.bar(bar_x,bar_y)
plt.show()
```

```python


# https://matplotlib.org/stable/gallery/color/named_colors.html#base-colors  doc제공 코드
import math
from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors


def plot_colortable(colors, *, ncols=4, sort_colors=True):

    cell_width = 212
    cell_height = 22
    swatch_width = 48
    margin = 12

    # Sort colors by hue, saturation, value and name.
    if sort_colors is True:
        names = sorted(
            colors, key=lambda c: tuple(mcolors.rgb_to_hsv(mcolors.to_rgb(c))))
    else:
        names = list(colors)

    n = len(names)
    nrows = math.ceil(n / ncols)

    width = cell_width * 4 + 2 * margin
    height = cell_height * nrows + 2 * margin
    dpi = 72

    fig, ax = plt.subplots(figsize=(width / dpi, height / dpi), dpi=dpi)
    fig.subplots_adjust(margin/width, margin/height,
                        (width-margin)/width, (height-margin)/height)
    ax.set_xlim(0, cell_width * 4)
    ax.set_ylim(cell_height * (nrows-0.5), -cell_height/2.)
    ax.yaxis.set_visible(False)
    ax.xaxis.set_visible(False)
    ax.set_axis_off()

    for i, name in enumerate(names):
        row = i % nrows
        col = i // nrows
        y = row * cell_height

        swatch_start_x = cell_width * col
        text_pos_x = cell_width * col + swatch_width + 7

        ax.text(text_pos_x, y, name, fontsize=14,
                horizontalalignment='left',
                verticalalignment='center')

        ax.add_patch(
            Rectangle(xy=(swatch_start_x, y-9), width=swatch_width,
                      height=18, facecolor=colors[name], edgecolor='0.7')
        )

    return fig
```

```python
plot_colortable(mcolors.CSS4_COLORS)
plt.show()
```

```python
# 해당하는 데이터를 발생지 시도별로 합계를 구하자
# 죽거나 다친 사람 = 사상자수
ctcp=ctc[['사상자수','발생지시도']].groupby('발생지시도').sum()
ctcp
```

```python
# bar 그래프를 그리기 위해서는 2차원 배열 --> 1차원 데이터로 변경해야함

y_bar = []  # -> 1차원 리스트
for i in range(len(y)) :  # len(y)는 17개
    y_bar.append(y[i][0])  #  y는 17행 1열


plt.figure(figsize=(15,8))
plt.bar(x,y_bar)
plt.xlabel('지역')
plt.ylabel('사상자수',rotation = 0)
plt.title('차대차 교통사고 사상자 수')
plt.show()
```

#  종합 연습문제 3 파이챠트 만들기
- 교통사고가 가장 많이 발생하는 시간대를 알아보고 시각화 해보자

```python
data.columns
```

```python
data['발생년월일시']
```

```python
data['발생년월일시'].astype('int64')
```

```python
# 발생년월일시에서 시간만 뽑자
time = data['발생년월일시'].astype('int64')
time %= 100
time
```

```python
a=2017123118
a%=100
a
```

```python
# 시간을 카테고리화 시켜보자
bins = [-1,2,5,8,11,14,17,20,23] # 시간구간의 끝숫자가 들어간다 # bins을 이용해 경계선을 잡는다 
# 예) 2는 경계선이고  -1도 경계선인데  0~2범위에 -1과2사이의 숫자인 0과1을 담고 경계선 중 2를 담는다 -1은 다음에 들어간다
# 0과 2까지 담아야 하므로 경계선인 -1을 적어주어야 한다
# 0을 경계선으로 하면 0은 0~2사이에 포함이 안된다.
# 경계선인 2는 0~2에 들어간다 5는 3~5에 들어간다 23은 21~23에 들어간다.근데 -1은 0~2에 들어가지 않는다.
label = [ '0~2','3~5','6~8','9~11','12~14','15~17','18~20','21~23']
rs = pd.cut(time,bins,labels=label)  # 잘라준다 -> 자르려고 하는 데이터를 보는 것이다
rs
```

```python
rs_data = rs.value_counts().sort_index() #- --> 인덱스 오름차순
#rs_data = rs.value_counts().sort_values() --> 값 오름차순
rs_data
```

```python
# 그래프 그리기
plt.figure(figsize=(5,5))
plt.pie(rs_data,labels=label,autopct='%1.1f%%') 
# 1. -> 점만찍었고 숫자가 없으므로 소숫점 없는 숫지
# 1.1 -> 점 찍고 1이므로 소수점 첫째자리까지
# 1.2 -> 점 찍고 2이므로 소수점 둘째자리까지
# %% 사이에 1.1f%을 넣는 것이다

plt.title('시간대별 사고 발생 비율')
plt.show()
```
