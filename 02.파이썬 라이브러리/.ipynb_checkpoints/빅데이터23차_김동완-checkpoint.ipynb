### 문제 3) 아래와 같이 리스트를 만들고 요소의 짝수/홀수를 구분하여 개수를 세는 프로그램을 작성하시오.

```python
list1 = [1, 50, 27, 19, 37, 33, 5 , 3 , 48]
a = 0
b = 0
for i in range(len(list1)) : 
    if list1[i]%2 == 0 :
        a+=1
    else :
        b+=1
print(f"짝수 {a}개, 홀수{b}개")
```

### 문제 4) 

```python
score_list = [['김미희', 89,78,99],
              ['서대희', 67,88,77],
              ['강예진', 100,55,96]]
print("="*50)
print("이름\tPython\tJava\tDB\t평균\t학점")
print("="*50)

for i in range(len(score_list)) :
    name = score_list[i][0]
    py = score_list[i][1]
    ja = score_list[i][2]
    db = score_list[i][3]
    mean1 = int((py+ja+db)/3)
    if mean1 >= 90 :
        ha = "A"
    elif mean1 >= 80 :
        ha = "B"
    elif mean1 >= 70 :
        ha = "C"
    elif mean1 >= 60 :
        ha = "D"
    else :
        ha = "F"
    print(f"{name}\t{py}\t{ja}\t{db}\t{mean1}\t{ha}")
print("="*50)
pysum = []
jasum = []
dbsum = []
for i in range(len(score_list)) :
    pysum.append(score_list[i][1])
    jasum.append(score_list[i][2])
    dbsum.append(score_list[i][3])
print(f"평균\t{int(sum(pysum)/len(pysum))}\t{int(sum(jasum)/len(jasum))}\t{int(sum(dbsum)/len(dbsum))}")
```

```python
data = pd.read_csv('./data/Traffic_Accident_2017.csv', encoding ='euc-kr')
data
```

```python
# 차대사람 빼오기
ctc = data[data['사고유형_대분류'] == '차대사람']
# 야간만 빼오기.
ab = ctc[ctc['주야'] == '야간']
```

```python
a = len(ab[ab["요일"]=="월"])
b = len(ab[ab["요일"]=="화"])
c = len(ab[ab["요일"]=="수"])
d = len(ab[ab["요일"]=="목"])
e = len(ab[ab["요일"]=="금"])
f = len(ab[ab["요일"]=="토"])
g = len(ab[ab["요일"]=="일"])
```

```python
temp = ab['요일'].value_counts()
y = temp[['월','화','수','목','금','토','일']]
y
```

```python
x = y.index
plt.title("요일별 야간 차대사람 교통사고 건수")
plt.bar(x,y)
plt.xlabel("요일")
plt.ylabel("사고건수")
plt.ylim(0,200)
plt.show()
```
